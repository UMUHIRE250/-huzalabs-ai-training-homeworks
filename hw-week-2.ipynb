**PART A - QUESTION 1  digitCount(n)

```python
def digitCount(n):
    # Initialize count to 0
    count = 0

    # Handle the case when n is 0 separately
    if n == 0:
        return 1

    # Take the absolute value of n to handle negative numbers
    n = abs(n)

    # Loop until n becomes 0
    while n != 0:
        # Remove the ones digit from n
        n //= 10
        # Increment the count of digits
        count += 1

    return count

# Test cases
print(digitCount(12323))  # Output: 5
print(digitCount(0))      # Output: 1
print(digitCount(-111))   # Output: 3
```

**PART A - QUESTION 2 gcd(m, n)

```python
def gcd(x, y):
    # Base case: if y is 0, then x is the gcd
    if y == 0:
        return x

    # Recursive case: apply Euclid's algorithm
    return gcd(y, x % y)

# Test cases
print(gcd(270, 250))  # Output: 10
```

Part A hasConsecutiveDigits(n)




```python
def hasConsecutiveDigits(n):
    # Convert n to a positive integer
    n = abs(n)

    # Initialize a variable to keep track of the previous digit
    prev_digit = None

    # Loop through each digit of the number
    while n > 0:
        # Get the last digit
        digit = n % 10

        # Check if the current digit is equal to the previous digit
        if digit == prev_digit:
            return True

        # Update prev_digit to the current digit
        prev_digit = digit

        # Remove the last digit from n
        n //= 10

    # If we reach this point, no consecutive digits were found
    return False

# Test cases
print(hasConsecutiveDigits(12323))  # Output: True
print(hasConsecutiveDigits(12345))  # Output: False
print(hasConsecutiveDigits(-111))   # Output: True
```

Part A Q4 nthAdditivePrime(n)

```python
def is_prime(num):
    if num <= 1:
        return False
    for i in range(2, int(num ** 0.5) + 1):
        if num % i == 0:
            return False
    return True

def sum_of_digits(n):
    return sum(int(digit) for digit in str(n))

def nthAdditivePrime(n):
    count = 0
    num = 2  # Start checking from 2
    while True:
        if is_prime(num) and is_prime(sum_of_digits(num)):
            count += 1
            if count == n:
                return num
        num += 1

# Test cases
print(nthAdditivePrime(1))  # Output: 2 (the 1st additive prime)
print(nthAdditivePrime(5))  # Output: 113 (the 5th additive prime)
```

Part A question 5 mostFrequentDigit(n)


```python
def mostFrequentDigit(n):
    # Initialize a dictionary to count the occurrences of each digit
    digit_count = {i: 0 for i in range(10)}

    # Convert n to positive to make it easier to work with digits
    n = abs(n)

    # Count the occurrences of each digit
    while n > 0:
        digit = n % 10
        digit_count[digit] += 1
        n //= 10

    # Find the maximum occurrence
    max_count = max(digit_count.values())

    # Find the smallest digit with maximum occurrence
    for i in range(9, -1, -1):
        if digit_count[i] == max_count:
            return i

# Test cases
print(mostFrequentDigit(12323))  # Output: 3
print(mostFrequentDigit(112233))  # Output: 1
print(mostFrequentDigit(-111))   # Output: 1
```

Part A question 6 isRotation(x, y)

```python
def isRotation(x, y):
    # Convert integers to strings for easy comparison of digits
    x_str = str(x)
    y_str = str(y)

    # Check if the lengths of x and y are equal
    if len(x_str) != len(y_str):
        return False

    # Concatenate y_str with itself to check for rotation
    y_str_rotated = y_str + y_str

    # Check if x_str is a substring of y_str_rotated
    if x_str in y_str_rotated:
        return True
    else:
        return False

# Test cases
print(isRotation(3412, 1234))  # Output: True
print(isRotation(1234, 3412))  # Output: True (rotation is symmetric)
print(isRotation(1234, 5678))  # Output: False
```

```python
def integral(f, a, b, N):
    # Step size
    h = (b - a) / N

    # Initialize the sum
    integral_sum = 0.5 * (f(a) + f(b))

    # Sum the function values at each subinterval
    for i in range(1, N):
        integral_sum += f(a + i * h)

    # Multiply by the step size
    integral_sum *= h

    return integral_sum

# Test function: f(x) = x^2
def f(x):
    return x**2

# Test the integral function
print(integral(f, 0, 1, 100))  # Output: approximately 0.3333333333333333
```

```python
def findZeroWithBisection(f, x0, x1, epsilon):
    # Ensure the initial interval contains a root
    if f(x0) * f(x1) > 0:
        raise ValueError("The initial interval does not contain a root.")

    # Perform bisection until the interval is sufficiently small
    while abs(x1 - x0) > epsilon:
        x_mid = (x0 + x1) / 2  # Calculate the midpoint of the interval
        if f(x_mid) == 0:  # Check if the midpoint is the root
            return x_mid
        elif f(x0) * f(x_mid) < 0:  # Check which subinterval to keep
            x1 = x_mid
        else:
            x0 = x_mid

    # Return the approximation of the root
    return (x0 + x1) / 2

# Example test function: f(x) = x^2 - 4
def f(x):
    return x**2 - 4

# Test the bisection method
approx_root = findZeroWithBisection(f, 1, 3, 0.0001)
print("Approximate root:", approx_root)
print("Function value at approximate root:", f(approx_root))
```

```python
def carrylessAdd(x, y):
    result = 0
    multiplier = 1  # To track the current place value

    # Iterate until both numbers become 0
    while x > 0 or y > 0:
        # Extract the rightmost digits of both numbers
        digit_x = x % 10
        digit_y = y % 10

        # Compute the sum of the digits (ignoring carries)
        digit_sum = (digit_x + digit_y) % 10

        # Update the result
        result += digit_sum * multiplier

        # Move to the next place value
        multiplier *= 10

        # Remove the rightmost digit from both numbers
        x //= 10
        y //= 10

    return result

# Test cases
print(carrylessAdd(785, 376))  # Output: 51
print(carrylessAdd(888, 999))   # Output: 867
```

Section B question 11

```python
def sum_of_digits(n):
    return sum(int(digit) for digit in str(n))

def prime_factors(n):
    factors = []
    divisor = 2
    while n > 1:
        while n % divisor == 0:
            factors.append(divisor)
            n //= divisor
        divisor += 1
    return factors

def nthSmithNumber(n):
    smith_count = 0
    number = 4  # Start from 4 since 1 is not considered a Smith number

    while True:
        if len(prime_factors(number)) > 1 and sum_of_digits(number) == sum(sum_of_digits(prime) for prime in prime_factors(number)):
            smith_count += 1
            if smith_count == n:
                return number
        number += 1

# Test cases
print(nthSmithNumber(1))  # Output: 4
print(nthSmithNumber(5))  # Output: 22
```

```python
import random

def roll_dice():
    return random.randint(1, 6)

def play_turn(player_name):
    total_points = 0
    while True:
        roll = roll_dice()
        print(f"{player_name} rolled a {roll}")
        if roll == 1:
            print("Oops! You rolled a 1. Your turn ends.")
            return 0
        else:
            total_points += roll
            choice = input("Do you want to roll again? (yes/no): ").lower()
            if choice != "yes":
                return total_points

def playPig():
    print("Welcome to the Pig game!")
    player1_name = input("Enter Player 1's name: ")
    player2_name = input("Enter Player 2's name: ")
    player1_points = 0
    player2_points = 0

    while player1_points < 100 and player2_points < 100:
        print("\n" + "-"*20)
        print(f"{player1_name}'s turn:")
        player1_points += play_turn(player1_name)
        print(f"{player1_name}'s total points: {player1_points}")

        if player1_points >= 100:
            print(f"Congratulations, {player1_name}! You won!")
            break

        print("\n" + "-"*20)
        print(f"{player2_name}'s turn:")
        player2_points += play_turn(player2_name)
        print(f"{player2_name}'s total points: {player2_points}")

        if player2_points >= 100:
            print(f"Congratulations, {player2_name}! You won!")
            break

# Start the game
playPig()
```

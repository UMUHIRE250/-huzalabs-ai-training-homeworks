```python
def rotateString(s, k):
    # If k is negative, convert it to positive
    if k < 0:
        k = len(s) + k

    # Perform rotation
    return s[k:] + s[:k]

# Test cases
print(rotateString('abcd',  1) == 'bcda') # output: 'bcda'
print(rotateString('abcd', -1) == 'dabc')

print("great")
```

```python
def applyCaesarCipher(message, shift):
    result = ""
    for char in message:
        # Check if the character is a letter
        if char.isalpha():
            # Determine the case (uppercase or lowercase)
            if char.isupper():
                base = ord('A')
            else:
                base = ord('a')

            # Apply the shift
            shifted_char = chr((ord(char) - base + shift) % 26 + base)
            result += shifted_char
        else:
            # If the character is not a letter, keep it unchanged
            result += char
    return result

# Test cases
print(applyCaesarCipher("We Attack At Dawn", 1) == "Xf Buubdl Bu Ebxo")
print(applyCaesarCipher("zodiac", -2) == "xmbgya")

print("Alright!")
```

```python
def largestNumber(text):
    numbers = []
    current_number = ""

    # Iterate through each character in the text
    for char in text:
        # If the character is a digit, add it to the current_number
        if char.isdigit():
            current_number += char
        elif current_number:
            # If current_number is not empty, convert it to an integer and add it to the numbers list
            numbers.append(int(current_number))
            current_number = ""  # Reset current_number

    # Add the last number if the text ends with a number
    if current_number:
        numbers.append(int(current_number))

    # Return the largest number if numbers list is not empty, otherwise return None
    return max(numbers) if numbers else None

# Test cases
print(largestNumber("I saw 3 dogs, 17 cats, and 14 cows!"))  # Output: 17
print(largestNumber("One person ate two hot dogs!"))         # Output: None
```

```python
def topScorer(data):
    # Split the data into lines
    lines = data.splitlines()

    # Initialize variables to keep track of the top score and corresponding player(s)
    top_score = 0
    top_scorers = []

    # Iterate through each line to calculate total scores
    for line in lines:
        # Split the line into player name and scores
        parts = line.split(',')
        player_name = parts[0]
        scores = [int(score) for score in parts[1:]]

        # Calculate the total score for the player
        total_score = sum(scores)

        # Update top score and top scorer(s) if needed
        if total_score > top_score:
            top_score = total_score
            top_scorers = [player_name]
        elif total_score == top_score:
            top_scorers.append(player_name)

    # Return the result
    if top_scorers:
        return ','.join(top_scorers)
    else:
        return None

# Test cases
data1 = '''\
Fred,10,20,30,40
Wilma,10,20,30
'''
print(topScorer(data1) == 'Fred')

data2 = '''\
Fred,10,20,30
Wilma,10,20,30,40
'''
print ( topScorer(data2) == 'Wilma')

data3 = '''\
Fred,11,20,30
Wilma,10,20,30,1
'''
assert topScorer(data3) == 'Fred,Wilma'

assert topScorer('') == None
```

```python
def collapseWhitespace(s):
    result = ''
    last_char_was_whitespace = False

    for char in s:
        if char.isspace():
            if not last_char_was_whitespace:
                result += ' '
                last_char_was_whitespace = True
        else:
            result += char
            last_char_was_whitespace = False

    return result.strip()

# Test cases
assert(collapseWhitespace("a\nb") == "a b")
assert(collapseWhitespace("a\n   \t    b") == "a b")
assert(collapseWhitespace("a\n   \t    b  \n\n  \t\t\t c   ") == "a b c")
print("all are well")
```

```python
def patternedMessage(message, pattern):
    # Remove leading and trailing whitespace from the pattern
    pattern = pattern.strip()

    result = ""
    message_index = 0

    # Iterate through each character in the pattern
    for char in pattern:
        if char.isspace():
            # Preserve whitespace characters in the pattern
            result += char
        else:
            # Replace non-whitespace characters with corresponding characters from the message
            if message_index < len(message):
                result += message[message_index]
                message_index += 1
            else:
                # If the message ends before the pattern, start over from the beginning of the message
                message_index = 0
                result += message[message_index]
                message_index += 1

    return result

# Test cases
print(patternedMessage("Go Pirates!!!", """
***************
******   ******
***************
"""))


# assert patternedMessage("Three Diamonds!","""
#     *     *     *
#    ***   ***   ***
#   ***** ***** *****
#    ***   ***   ***
#     *     *     *
# """) == 'ThreeDiamonds!\ns!ThreeDi am ond\ns!Three Diamonds!\n   !Th ree   Di\n    mon'
```

```python
import math

def encodeRightLeftRouteCipher(message, rows):
    # Determine the number of columns needed in the grid
    columns = math.ceil(len(message) / rows)

    # Calculate the number of characters needed to pad the message
    padding_length = rows * columns - len(message)

    # Pad the message with lowercase letters starting from 'z'
    padded_message = message + ''.join(chr(122 - i) for i in range(padding_length))

    # Initialize the encoded message with the number of rows
    encoded_message = str(rows)

    # Loop through each row in the grid
    for row in range(rows):
        # Determine the direction of reading for the current row
        if row % 2 == 0:
            # Left-to-right reading for even rows
            encoded_message += ''.join(padded_message[row + columns * col] for col in range(columns))
        else:
            # Right-to-left reading for odd rows
            encoded_message += ''.join(padded_message[(row + 1) * columns - col - 1] for col in range(columns))

        print("Row:", row)
        print("Encoded Message:", encoded_message)

    return encoded_message

# Test cases
assert encodeRightLeftRouteCipher("WEATTACKATDAWN", 4) == "4WTAWNTAEACDzyAKT"
assert encodeRightLeftRouteCipher("WEATTACKATDAWN", 3) == "3WTCTWNDKTEAAAAz"
assert encodeRightLeftRouteCipher("WEATTACKATDAWN", 5) == "5WADACEAKWNATTTz"
```
